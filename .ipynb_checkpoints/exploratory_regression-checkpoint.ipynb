# REGRESSION

```python
def calculate_vif(df):
    # Add constant term for intercept
    X = sm.add_constant(df)
    
    vif_data = pd.DataFrame()
    vif_data["feature"] = X.columns
    vif_data["VIF"] = [variance_inflation_factor(X.values, i) for i in range(X.shape[1])]
    
    # Remove intercept row from VIF report
    return vif_data[vif_data['feature'] != 'const']

def remove_high_vif_features(df, threshold=10.0):
    df_clean = df.copy()
    while True:
        vif = calculate_vif(df_clean)
        max_vif = vif['VIF'].max()
        if max_vif > threshold:
            # Drop the feature with the highest VIF
            feature_to_drop = vif.sort_values('VIF', ascending=False)['feature'].iloc[0]
            print(f"Dropping '{feature_to_drop}' with VIF={max_vif:.2f}")
            df_clean = df_clean.drop(columns=[feature_to_drop])
        else:
            break
    return df_clean
```

```python
def lasso_and_plot(X,y, lasso=True):
    # Split and scale
    X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=42)
    scaler = StandardScaler()
    X_train_scaled = scaler.fit_transform(X_train)
    X_test_scaled = scaler.transform(X_test)

    if lasso:
        # LassoCV to find important features
        lasso_cv = LassoCV(cv=5, random_state=42)
        lasso_cv.fit(X_train_scaled, y_train)
        
        # Get mask for selected features (non-zero coef)
        selected_mask = lasso_cv.coef_ != 0
        selected_features = X.columns[selected_mask]
        print("Selected features:", list(selected_features))

    else:
        selected_features = X.columns
    
    # Prepare data for statsmodels OLS (only selected features)
    X_train_sel = X_train.loc[:, selected_features]
    X_train_sel = sm.add_constant(X_train_sel)  # add intercept
    
    # Fit OLS model to get p-values and coef
    ols_model = sm.OLS(y_train, X_train_sel).fit()
    
    print(ols_model.summary())
    
    # Plot coefficients and p-values
    coef = ols_model.params[1:]  # skip intercept
    pvals = ols_model.pvalues[1:]
    conf_int = ols_model.conf_int().iloc[1:]  # skip intercept row
    
    lower = conf_int[0]
    upper = conf_int[1]
    
    colors = ['#3288bd' if p <= 0.05 else '#fc8d59' for p in pvals]
    
    fig, ax = plt.subplots(figsize=(8, 0.3*len(pvals)))
    
    # Plot coefficients as bars
    ax.barh(coef.index, coef.values, xerr=[coef - lower, upper - coef], color=colors, alpha=0.8, capsize=5)
    
    ax.set_xlabel('Coefficient')
    ax.set_ylabel('Features')
    
    # Legend for significance
    import matplotlib.patches as mpatches
    sig_patch = mpatches.Patch(color='#3288bd', label='p ≤ 0.05 (significant)')
    nonsig_patch = mpatches.Patch(color='#fc8d59', label='p > 0.05 (not significant)')
    ax.legend(handles=[sig_patch, nonsig_patch], loc='best')
    
    plt.title('Coefficients with 95% Confidence Intervals')
    plt.show()

    return ols_model
```

```python
# Select features 
X = df_all[columns_all]
X = X.astype(float)
# Target variable
y = df_all['percentage_watched']

# Example usage:
# Assuming X is your dataframe of features (excluding target)
X_clean = remove_high_vif_features(X, threshold=10)

print("Remaining features after VIF-based cleaning:")
print(X_clean.columns)
```

```python
print(X.shape, X_clean.shape)
```

## With LASSO

```python
model_lasso = lasso_and_plot(X_clean,y)
```

```python
# Filter by p-value
summary_table_lasso = model_lasso.summary2().tables[1]
significant_features_lasso = summary_table_lasso[summary_table_lasso['P>|t|'] < 0.05]
df_features_lasso = pd.DataFrame(significant_features_lasso.sort_values('Coef.'))
df_features_lasso
```

```python
print(X.shape, X_clean.shape, df_features_lasso.shape)
```

```python
# Get the summary table
summary_table = model.summary2().tables[1]

# Drop intercept row
coef_df = summary_table.drop('const', errors='ignore').copy()

# Sort by coefficient size for better visualization (optional)
coef_df = coef_df.sort_values(by='Coef.')

# Prepare plot data
features = coef_df.index
coefficients = coef_df['Coef.']
conf_lower = coef_df['[0.025']
conf_upper = coef_df['0.975]']

# Plot
plt.figure(figsize=(10, len(features)*0.3))
plt.errorbar(coefficients, features, xerr=[coefficients - conf_lower, conf_upper - coefficients],
             fmt='s', color='black', ecolor='gray', elinewidth=3, capsize=4)

plt.axvline(x=0, color='red', linestyle='--')
plt.xlabel('Coefficient Estimate')
plt.title('Linear Regression Coefficients with 95% Confidence Intervals')
plt.tight_layout()
plt.show()
```
